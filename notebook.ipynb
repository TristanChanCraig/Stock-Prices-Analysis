## Stock Market Data Visualization

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Task 1: Unemployment Rate over GDP

```python
df = pd.read_csv('/workspaces/Stock-Prices-Analysis/data.csv')

plt.figure(figsize=(12, 8))

info = df.groupby('Unemployment_Rate')['GDP'].sum()
info.plot(kind='line', color='black')
plt.title('GDP (billions) over Unemployment Rate')
plt.ylabel('GDP (billions)')
plt.xlabel('Unemployment Rate')
plt.xticks(rotation=90)
plt.show()
```

The GDP reached its peak around 3% unemployment rate at 33 trillion and rapidly fell soon after to around 5 trillion at higher rates of unemployment.

## Task 2: CPI over PPI

```python
# xlim=(0, 10), ylim=(0, 10)
plt.figure(figsize=(12, 8), dpi=600)
info = df.groupby('CPI')['PPI'].sum()
info.plot(kind='line', color='black')
plt.title('PPI over CPI')
plt.ylabel('PPI')
plt.xlabel('CPI')
plt.xticks(rotation=90)
plt.show()
```

## Task 3: Day of the Week over Volume

```python
info = df.groupby('DayOfWeek')['Volume'].sum().plot(kind='barh', color='red')
plt.title('Day of Week over Volume')
plt.ylabel('Day of Week')
plt.xlabel('Volume')
plt.xticks(rotation=90)
plt.show()
```

## Task 4: Open over Close

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Open', y='Close/Last', data=df)
plt.title('Close over Open')
plt.ylabel('Close')
plt.xlabel('Open')
plt.show()
```
